```python
from keras.preprocessing.text import Tokenizer
```

```python
def create_tokenizer(num_words, lower, oov_token, texts, mode):

  # Construct tokenizer
  tokenizer = Tokenizer(num_words = num_words, lower = lower, oov_token = oov_token)
  
  # Build vocabulary
  tokenizer.fit_on_texts(texts = texts)

  # Convert and return integer vectors
  return tokenizer.texts_to_matrix(texts = texts, mode = mode)
```

```python
texts = [
  'Hello, I am Kevin.',
  'Hello kevin. I am Alex.',
  'How are you Alex?',
  'I am doing great, how are you Kevin?',
  'Awesome!'
]

matrix = create_tokenizer(num_words = 10, lower = True, oov_token = '?', texts = texts, mode = 'freq')
matrix
```
